```python
def mannwhitneyu_test(df5, df20, output_csa_name, propose):
    column_names = ["algorithm", "dim5", "dim20"]
    o = pd.DataFrame(columns = column_names)
    if propose == 'SEP_fix_VS_UNIOA':
        o['algorithm'] = ['BA', 'GOA', 'CSA', 'MFO', 'MBO', 'BOA', 'PSO']
        for col in ["dim5", "dim20"]:
            if col == "dim5":
                df = df5
            else:
                df = df20
            ba = stats.ttest_rel(df['SEP_BA_fix'], df['UNIOA_BA'])[1]
            goa = stats.ttest_rel(df['SEP_GOA'], df['UNIOA_GOA'])[1]
            csa = stats.ttest_rel(df['SEP_CSA'], df['UNIOA_CSA'])[1]
            mfo = stats.ttest_rel(df['SEP_MFO'], df['UNIOA_MFO'])[1]
            mbo = stats.ttest_rel(df['SEP_MBO'], df['UNIOA_MBO'])[1]
            boa = stats.ttest_rel(df['SEP_BOA_fix'], df['UNIOA_BOA'])[1]
            pso = stats.ttest_rel(df['SEP_PSO_fix'], df['UNIOA_PSO'])[1]
            o[col] = [ba, goa, csa, mfo, mbo, boa, pso]
            
    if propose == 'syncE_VS_asyncE':
        o['algorithm'] = ['UNIOA_BA_syncE_or_asyncE_syncG', 'UNIOA_GOA_syncE_or_asyncE_syncG', 'UNIOA_CSA_syncE_or_asyncE_syncG',
                          'UNIOA_MFO_syncE_or_asyncE_syncG', 'UNIOA_MBO_syncE_or_asyncE_syncG', 'UNIOA_BOA_syncE_or_asyncE_syncG',
                          'UNIOA_PSO_syncE_or_asyncE_syncG',
                          'SEP_BA_syncE_or_asyncE_syncG', 'SEP_GOA_syncE_or_asyncE_syncG', 'SEP_CSA_syncE_or_asyncE_syncG',
                          'SEP_MFO_syncE_or_asyncE_syncG', 'SEP_MBO_syncE_or_asyncE_syncG', 'SEP_BOA_syncE_or_asyncE_syncG',
                          'SEP_PSO_syncE_or_asyncE_syncG']
        for col in ["dim5", "dim20"]:
            if col == "dim5":
                df = df5
            else:
                df = df20
            ba  = stats.ttest_rel(df['UNIOA_BA_syncE_syncG'],  df['UNIOA_BA_asyncE_syncG'])[1]
            goa = stats.ttest_rel(df['UNIOA_GOA_syncE_syncG'], df['UNIOA_GOA_asyncE_syncG'])[1]
            csa = stats.ttest_rel(df['UNIOA_CSA_syncE_syncG'], df['UNIOA_CSA_asyncE_syncG'])[1]
            mfo = stats.ttest_rel(df['UNIOA_MFO_syncE_syncG'], df['UNIOA_MFO_asyncE_syncG'])[1]
            mbo = stats.ttest_rel(df['UNIOA_MBO_syncE_syncG'], df['UNIOA_MBO_asyncE_syncG'])[1]
            boa = stats.ttest_rel(df['UNIOA_BOA_syncE_syncG'], df['UNIOA_BOA_asyncE_syncG'])[1]
            pso = stats.ttest_rel(df['UNIOA_PSO_syncE_syncG'], df['UNIOA_PSO_asyncE_syncG'])[1]
            
            ba1  = stats.ttest_rel(df['SEP_BA_syncE_syncG'],  df['SEP_BA_asyncE_syncG'])[1]
            goa1 = stats.ttest_rel(df['SEP_GOA_syncE_syncG'], df['SEP_GOA_asyncE_syncG'])[1]
            csa1 = stats.ttest_rel(df['SEP_CSA_syncE_syncG'], df['SEP_CSA_asyncE_syncG'])[1]
            mfo1 = stats.ttest_rel(df['SEP_MFO_syncE_syncG'], df['SEP_MFO_asyncE_syncG'])[1]
            mbo1 = stats.ttest_rel(df['SEP_MBO_syncE_syncG'], df['SEP_MBO_asyncE_syncG'])[1]
            boa1 = stats.ttest_rel(df['SEP_BOA_syncE_syncG'], df['SEP_BOA_asyncE_syncG'])[1]
            pso1 = stats.ttest_rel(df['SEP_PSO_syncE_syncG'], df['SEP_PSO_asyncE_syncG'])[1]
            
            o[col] = [ba, goa, csa, mfo, mbo, boa, pso,
                     ba1, goa1, csa1, mfo1, mbo1, boa1, pso1,]
            
    if propose == 'syncG_VS_asyncG':
        o['algorithm'] = ['SEP_ba', 'SEP_goa', 'SEP_csa', 'SEP_mfo', 'SEP_mbo', 'SEP_boa', 'SEP_pso',
                         'UNIOA_ba', 'UNIOA_goa', 'UNIOA_csa', 'UNIOA_mfo', 'UNIOA_mbo', 'UNIOA_boa', 'UNIOA_pso']
        for col in ["dim5", "dim20"]:
            if col == "dim5":
                df = df5
            else:
                df = df20
            ba = stats.ttest_rel(df['SEP_BA_fix'], df['UNIOA_BA'])[1]
            goa = stats.ttest_rel(df['SEP_GOA'], df['UNIOA_GOA'])[1]
            mbo = stats.ttest_rel(df['SEP_MBO'], df['UNIOA_MBO'])[1]
            boa = stats.ttest_rel(df['SEP_BOA_fix'], df['UNIOA_BOA'])[1]
            pso = stats.ttest_rel(df['SEP_PSO_fix'], df['UNIOA_PSO'])[1]
            o[col] = [ba, goa, mbo, boa, pso]
    
    return o.to_csv(output_csa_name + '.csv', index=False)
```
